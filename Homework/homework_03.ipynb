```python
def count_word_frequency(text):
    """Возвращает словарь: слово -> количество вхождений в тексте.

    Текст приводится к нижнему регистру, простые знаки препинания
    (точка, запятая, восклицательный и вопросительный знак) удаляются.
    """
    # Приводим текст к нижнему регистру
    text = text.lower()

    # Удаляем простые знаки препинания
    for znak in [".", ",", "!", "?"]:
        text = text.replace(znak, "")

    # Разбиваем строку на слова по пробелам
    words = text.split()

    # Считаем частоту появления каждого слова
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1

    return frequency
```
