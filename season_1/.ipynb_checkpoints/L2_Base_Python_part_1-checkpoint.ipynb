# Зачем нам Git? 


Представьте, что вы пишете важный отчет или курсовую работу. Как выглядит ваша папка с файлами? Вероятно, как-то так:

- `Отчет.docx`
- `Отчет_правки.docx`
- `Отчет_после_замечаний_научного_руководителя.docx`
- `Отчет_ФИНАЛ.docx`
- `Отчет_ФИНАЛ_v2_ТОЧНО_ПОСЛЕДНИЙ.docx`

Это хаос. Что если вы удалили что-то важное в `v2` и хотите вернуться к версии `после_замечаний`? Что если ваш коллега вносил правки в свою копию, а вы — в свою, и теперь нужно их как-то объединить? Эта проблема стара как мир, и для ее решения были созданы **Системы Контроля Версий (Version Control Systems, VCS)**.

Самая популярная такая система сегодня — это **Git**.

**Git** — это программа для управления этими "сохранениями" на вашем компьютере.
**GitHub** (а также GitLab, Bitbucket) — это веб-сайт, "облако" для хранения ваших проектов с Git. Это как Google Drive для кода.


**Нам это нужно для двух вещей:**
1.  **Сдача ДЗ:** Вы будете выкладывать свои тетрадки в публичный репозиторий на GitHub, а мне присылать ссылку.
2.  **Знание:** Ваш GitHub — это ваше будущее портфолио. Работодатели смотрят на него в первую очередь.

## Аналогия: Git — это "сохранения" в вашей работе

Давайте забудем о сложном определении и представим, что ваша работа над проектом — это прохождение большой и сложной игры.

- **Репозиторий (Repository):** Это папка с вашей "игрой" (вашим проектом).
- **Коммит (Commit):** Это точка сохранения. Вы дошли до важного места (написали работающую функцию, закончили главу отчета) и нажимаете "Сохранить игру". Каждому сохранению вы даете осмысленное название, например, "Победил первого босса" или "Добавил расчет среднего чека". Теперь, если что-то пойдет не так, вы всегда сможете "загрузиться" с этого сохранения.
- **Ветка (Branch):** Представьте, что вы хотите поэкспериментировать. Например, попробовать пройти уровень с новым мечом, не теряя своего старого сохранения. Вы создаете "альтернативную временную линию" — это и есть новая ветка. Вы можете свободно экспериментировать в ней, и если эксперимент удался — вы можете "слить" эту временную линию с основной. Если нет — просто удаляете ее.

**Git** — это и есть движок этой "игры", который управляет всеми сохранениями и временными линиями.

## Git vs. GitHub: В чем разница?

Это самый частый источник путаницы для новичков. Давайте проясним раз и навсегда.

**Git**
- Это **программа**, инструмент, который устанавливается на ваш компьютер.
- Он работает локально, в командной строке.
- Его задача — отслеживать изменения в файлах, создавать "коммиты", управлять "ветками".
- **Аналогия:** Это как Microsoft Word. Программа для написания текстов, которая живет на вашем компьютере.

**GitHub**
- Это **веб-сайт**, сервис, облачная платформа.
- Он предоставляет "место в облаке" для хранения ваших проектов (репозиториев).
- Он добавляет социальные функции: вы можете делиться кодом, работать в команде, комментировать чужой код.
- **Аналогия:** Это как Google Docs или Dropbox. Сервис для хранения ваших документов в интернете и совместной работы над ними.

| | Git | GitHub |
| :--- | :--- | :--- |
| **Что это?** | Программа (инструмент) | Веб-сайт (сервис) |
| **Где работает?** | Локально на вашем ПК | В облаке (в интернете) |
| **Основная задача**| Контроль версий | Хостинг репозиториев и коллаборация |

**Вывод:** Мы используем **Git**, чтобы управлять историей нашего проекта на компьютере, а затем отправляем эту историю на **GitHub**, чтобы сохранить резервную копию и поделиться ею с другими (в нашем случае — со мной для проверки ДЗ).

## Почему это так важно для аналитика данных?

Может показаться, что это инструмент чисто для программистов. Это не так. Для специалиста по данным Git/GitHub — это три кита профессионализма:

1.  **Воспроизводимость и порядок.** Анализ данных — это исследование. Вы пробуете разные гипотезы, меняете код, добавляете новые графики. Git позволяет вам отслеживать ход ваших мыслей. Вы всегда можете вернуться к версии, которая дала интересный результат, и вам не нужно хранить 10 копий одной тетрадки.

2.  **Командная работа.** Редкий аналитический проект делается в одиночку. Вы будете работать вместе с другими аналитиками, инженерами данных, менеджерами. Git — это индустриальный стандарт для совместной работы над кодом. Без него вы не сможете эффективно работать в команде.

3.  **Ваше публичное портфолио.**  Ваш профиль на GitHub — это **ваше  резюме**. Когда работодатель видит активный профиль с учебными проектами, домашними заданиями, он понимает:
    - Этот человек знаком с базовыми инструментами разработки.
    - Можно посмотреть на стиль его кода и ход мыслей.

# Что такое "Операторы" и "Выражения"?

В прошлой лекции мы уже интуитивно использовали эти понятия. Давайте теперь дадим им четкие определения.

- **Выражение (Expression):** Это любой кусочек кода, который **"вычисляется"** и в результате дает какое-то **значение**.
    - `2 + 2` — это выражение. Его значение — `4`.
    - `"Привет" + " " + "мир"` — это выражение. Его значение — `"Привет мир"`.
    - `temperature > 20` — это выражение. Его значение будет `True` или `False`.

- **Оператор (Operator):** Это специальный символ (или символы), который выполняет операцию над одним или несколькими значениями (операндами).
    - В выражении `2 + 2`, `2` — это операнды, а `+` — это оператор.

- **Инструкция (Statement):** Это целая "команда" или "предложение" на языке Python, которое выполняет какое-то действие. Инструкции часто содержат выражения.
    - `total_cost = 250 * 2` — это инструкция. Она содержит выражение `250 * 2`, вычисляет его значение (`500`) и присваивает его переменной `total_cost`.

Проще говоря, мы **составляем выражения из операндов и операторов**, чтобы **написать инструкцию**, которую выполнит Python.

##  Арифметические операторы: Углубленный взгляд

С большинством из них мы уже знакомы. Давайте систематизируем и добавим несколько новых, очень полезных.

````python
# --- Знакомые операторы ---
print(f"Сложение: 10 + 3 = {10 + 3}")
print(f"Вычитание: 10 - 3 = {10 - 3}")
print(f"Умножение: 10 * 3 = {10 * 3}")
print(f"Обычное деление (результат всегда float): 10 / 3 = {10 / 3}")
print("-" * 20)

# --- Новые и важные операторы ---

# 1. Целочисленное деление (Floor Division) //
# Показывает, сколько ЦЕЛЫХ раз одно число помещается в другое. Дробная часть отбрасывается.
bananas = 25
people = 7
print(f"У нас {bananas} бананов и {people} человек.")
print(f"Каждый получит по {bananas // people} банана.")

# 2. Остаток от деления (Modulo) %
# Показывает, что останется после целочисленного деления.
print(f"А {bananas % people} банана останется в качестве 'налога для организатора' :)")

# Практическое применение %: проверка на четность
# Если число % 2 == 0, оно четное. Иначе - нечетное. Это используется ПОСТОЯННО.
number_to_check = 14
if number_to_check % 2 == 0:
    print(f"Число {number_to_check} - четное.")
else:
    print(f"Число {number_to_check} - нечетное.")
    
# 3. Возведение в степень (Exponentiation) **
print(f"\n2 в 10 степени = {2**10}")
print(f"Квадратный корень из 9 (это 9 в степени 0.5) = {9**0.5}")
````

##  Операторы присваивания: Упрощаем себе жизнь

Мы уже знаем главный оператор присваивания `=`. Но есть и его "старшие братья", которые делают код короче и читабельнее.

````python
# Базовое присваивание
balance = 1000
print(f"Начальный баланс: {balance}")

# Мы хотим пополнить баланс на 500
# Длинный способ:
balance = balance + 500
print(f"Баланс после пополнения (длинный способ): {balance}")

# Короткий и предпочтительный способ с оператором +=
balance += 500  # Это полностью эквивалентно balance = balance + 500
print(f"Баланс после пополнения (короткий способ): {balance}")

# Такие же сокращения есть для всех арифметических операторов:
balance -= 300  # Списание
print(f"Баланс после списания 300: {balance}")

balance *= 1.1 # Начисление процентов (увеличение на 10%)
print(f"Баланс после начисления 10% годовых: {balance:.2f}")

balance /= 2 # Разделили деньги с другом
print(f"Баланс после дележки: {balance:.2f}")
````

## Методы: Встроенные "умения" объектов

В Python почти всё, с чем мы работаем (числа, строки, списки), — это **объекты**.

**Аналогия:** Представьте, что строка — это не просто кусок текста, а "умный" объект, у которого есть свои встроенные "умения" или "навыки". Например, у строки есть навык "превратить себя в заглавные буквы" или "посчитать, сколько раз в ней встречается определенный символ".

Эти встроенные умения называются **методами**. Мы "вызываем" метод через точку после имени переменной.

**Синтаксис:** `имя_переменной.имя_метода(аргументы)`

````python
# Давайте создадим объект-строку
my_string = "   Привет, МИСИС! Привет, Python!   "
print(f"Оригинал строки: '{my_string}'")
print("-" * 30)
````

````python
# Используем (вызываем) его методы:

# .upper() - возвращает НОВУЮ строку в верхнем регистре
upper_string = my_string.upper()
print(f"Метод .upper(): {upper_string}")
````

````python
# .lower() - возвращает НОВУЮ строку в нижнем регистре
lower_string = my_string.lower()
print(f"Метод .lower(): {lower_string}")
````

````python
# .strip() - возвращает НОВУЮ строку, убрав лишние пробелы в начале и в конце. Очень полезно при очистке данных!
stripped_string = my_string.strip()
print(f"Метод .strip(): '{stripped_string}'")
````

````python
# .replace(что_меняем, на_что_меняем) - заменяет все вхождения одной подстроки на другую
replaced_string = stripped_string.replace("Привет", "Пока")
print(f"Метод .replace(): '{replaced_string}'")
````

````python
# .count(что_считаем) - считает количество вхождений подстроки
count_privet = stripped_string.count("Привет")
print(f"Метод .count('Привет'): {count_privet}")
````

````python
# .title() - делает заглавной первую букву каждого слова
title_string = "иван иванович иванов"
print(f"\nМетод .title() для '{title_string}': '{title_string.title()}'")
````

````python
# .startswith(подстрока) - проверяет, начинается ли строка с указанной подстроки (возвращает True/False)
filename = "report_2024.csv"
print(f"Файл '{filename}' начинается с 'report_'? {filename.startswith('report_')}")
print(f"Файл '{filename}' заканчивается на '.csv'? {filename.endswith('.csv')}")
````


Все методы, название которых начинается с двух подчёркиваний являются служебными. Это не значит, что их нельзя вызывать напрямую, надо просто иметь в виду что они будут вызываться, когда интерпретатор увидит соответствующий оператор. 
Это удобно, потому что их можно перегружать для своих кастомных классов

````python
a = 15
print(a + 5)
print(a.__add__(5))
print(a < 10)
print(a.__lt__(10))  # lt от less than
````

**❗️Важный момент: Иммутабельность строк**

Обратите внимание, что методы строк **не изменяют исходную строку**. Они **возвращают новую**, измененную копию. Такой тип данных называется **иммутабельным** (неизменяемым).

Это как с фотографией: вы можете отфотошопить ее и сохранить как новый файл, но исходный снимок останется нетронутым.

````python
original_phrase = "Я люблю спать и кушать"
print(f"Исходная строка (до методов): {original_phrase}")

# Вызываем метод, но не сохраняем результат в переменную
original_phrase.upper()

print(f"Исходная строка (после вызова .upper()): {original_phrase}") # Ничего не изменилось!

# Чтобы изменение "сохранилось", нужно перезаписать переменную
original_phrase = original_phrase.upper()
print(f"Исходная строка (после перезаписи): {original_phrase}")
````

#  Переменные: Глубокое погружение

Мы уже знаем, что переменная — это "коробка с этикеткой" для хранения данных. Тоесть Переменная это такая штука (число/строка/объект), которую мы обозначаем особой меткой.


> Можно называть переменные русскими буквами. И если вы так будете делать, вы получите ачивку **"У меня нет друзей"**

## Правила именования и стиль кода (PEP 8)

Как вы назовете переменную, имеет огромное значение. Хорошее имя — залог читаемости вашего кода. В мире Python существует официальный "свод правил хорошего тона" — **PEP 8 (Python Enhancement Proposal #8)**.

**Ключевые правила из PEP 8 для именования переменных:**

1.  **Используйте `snake_case`:** Имена переменных должны быть в нижнем регистре, а слова разделяться нижним подчеркиванием.
    - **Хорошо:** `user_name`, `order_total_price`, `is_active`.
    - **Плохо:** `UserName`, `ordertotalprice`, `isActive` (последнее — стиль `camelCase` из других языков, в Python он не принят для переменных).

2.  **Имя должно быть осмысленным:** Имя переменной должно описывать то, что в ней хранится.
    - **Хорошо:** `user_age = 30`
    - **Плохо:** `x = 30` (Что такое `x`? Непонятно без дополнительного контекста).

## Ловушка `input()`: Всё есть строка!

Это **самое важное правило** и самая частая ошибка новичков. Запомните его навсегда:

> **Функция `input()` ВСЕГДА возвращает значение типа СТРОКА (`str`), даже если пользователь ввел одни цифры.**

Попытка использовать результат `input()` в математических операциях приведет к ошибке.

````python
# Попытка написать калькулятор
print("--- Простой калькулятор сложения ---")
num1_str = input("Введите первое число: ")
num2_str = input("Введите второе число: ")

print(f"Тип первого 'числа': {type(num1_str)}")
print(f"Тип второго 'числа': {type(num2_str)}")

# Что произойдет, если мы сложим две строки? Конкатенация!
result_wrong = num1_str + num2_str
print(f"Результат сложения строк: {result_wrong}") # Если ввели 5 и 7, получим "57"

# А если попробуем умножить? Ошибка!
# result_error = num1_str * num2_str # TypeError: can't multiply sequence by non-int of type 'str'
````

## Решение: Явное преобразование типов (Type Casting)

Чтобы решить эту проблему, нам нужно явно сказать Python: "Эй, я знаю, что это строка, но я уверен, что внутри нее число. Пожалуйста, преврати ее в числовой тип".

Для этого существуют встроенные функции-конвертеры:
- `int(x)`: пытается превратить `x` в целое число.
- `float(x)`: пытается превратить `x` в число с плавающей точкой.
- `str(x)`: пытается превратить `x` в строку (это мы уже использовали ранее).

Обычно `input()` сразу "оборачивают" в нужную функцию-конвертер.

````python
# Правильный калькулятор
print("--- Правильный калькулятор сложения ---")

# Сразу преобразуем результат input в нужный тип
num1 = float(input("Введите первое число: ")) # Используем float, чтобы можно было вводить дробные числа
num2 = float(input("Введите второе число: "))

print(f"Тип первого числа: {type(num1)}")
print(f"Тип второго числа: {type(num2)}")

result_correct = num1 + num2
print(f"Результат правильного сложения: {result_correct}")
````

# Введение: Как научить программу "думать"

До сих пор наши программы были как повара, строго следующие рецепту: "возьми это, положи то, смешай". Они не могли принимать никаких решений. Но что, если нам нужно, чтобы программа вела себя по-разному в зависимости от ситуации?
- "Если пользователь совершеннолетний, разрешить доступ, **иначе** — запретить".
- "Если температура ниже нуля **и** идет снег, включить отопление".
- "Если пользователь ввел "выход" **или** нажал `Esc`, завершить программу".

Чтобы реализовать такую логику, нам нужны инструменты для сравнения и комбинирования условий. Этими инструментами и являются **операторы сравнения** и **логические операторы**. Они всегда возвращают один из двух возможных ответов: `True` (Истина) или `False` (Ложь).

## Операторы сравнения: Основа всех решений

Операторы сравнения — это базовые "вопросы", которые мы можем задать о данных. Они берут два значения и сравнивают их.

````python
# --- Основные операторы сравнения ---

# == (Равно). Внимание! Не путать с = (присваивание)!
# Это самый частый источник ошибок у новичков.
# =  -> "Положить значение в коробку"
# == -> "Сравнить, равны ли значения в двух коробках"
print(f"5 == 5? -> {5 == 5}")
print(f"5 == 6? -> {5 == 6}")

# != (Не равно)
print(f"5 != 6? -> {5 != 6}")

# > (Больше)
print(f"5 > 3? -> {5 > 3}")

# < (Меньше)
print(f"5 < 3? -> {5 < 3}")

# >= (Больше или равно)
print(f"5 >= 5? -> {5 >= 5}")

# <= (Меньше или равно)
print(f"5 <= 4? -> {5 <= 4}")
````

  "Математические" операторы:

    == "равно"
    != "не равно"
    > (левая часть больше, чем правая)
    >= (левая часть больше, либо равна правой)
    <  (левая часть меньше, чем правая)
    <= (левая часть меньше, либо равна правой)

Естественно, сравнивать можно не только числа, но и другие типы данных. Однако, если с равенством\неравенством ещё интуитивно понятно, то вот > и < для строк и списков не интуитивная операция, частично мы затронем её дальше.

### Сравнение разных типов данных

Сравнивать можно не только числа.

````python
# Сравнение строк
# Сравнение происходит лексикографически (по алфавиту, как в словаре).
# Учитывается регистр: заглавные буквы "меньше" строчных.
print(f"'яблоко' == 'яблоко'? -> {'яблоко' == 'яблоко'}")
print(f"'яблоко' < 'банан'? -> {'яблоко' < 'банан'}") # False, 'я' идет после 'б'
print(f"'Python' == 'python'? -> {'Python' == 'python'}") # False, регистр разный

# Сравнение с переменными
user_password = "password123"
entered_password = input("Введите пароль: ")
is_password_correct = (user_password == entered_password)
print(f"Пароль верный? -> {is_password_correct}")
````

## Логические операторы: `and`, `or`, `not`

Если операторы сравнения — это простые вопросы, то логические операторы — это "союзы", которые позволяют нам строить из них сложные предложения.

### Оператор `and` (логическое "И")

Выражение `A and B` истинно (`True`) только в том случае, если **ОБА** `A` и `B` истинны. Если хотя бы одно из них ложно, все выражение становится ложным.

| A | B | A and B |
| :---: | :---: | :---: |
| `True` | `True` | `True` |
| `True` | `False` | `False` |
| `False`| `True` | `False` |
| `False`| `False`| `False` |

**Аналогия:** Чтобы получить водительские права, вам должно быть `(возраст >= 18) and (вы немного адекватны  - есть медсправка)`. Если хотя бы одно условие не выполнено, прав вы не получите.

````python
age = 25
has_driver_license = True
can_rent_a_car = (age >= 21) and (has_driver_license == True)
print(f"Возраст {age}, права есть. Можно арендовать машину? -> {can_rent_a_car}")

age = 19
has_driver_license = True
can_rent_a_car = (age >= 21) and (has_driver_license == True)
print(f"Возраст {age}, права есть. Можно арендовать машину? -> {can_rent_a_car}")
````

### Оператор `or` (логическое "ИЛИ")

Выражение `A or B` истинно (`True`), если **ХОТЯ БЫ ОДНО** из выражений `A` или `B` истинно. Оно ложно только тогда, когда оба выражения ложны.

| A | B | A or B |
| :---: | :---: | :---: |
| `True` | `True` | `True` |
| `True` | `False` | `True` |
| `False`| `True` | `True` |
| `False`| `False`| `False` |

**Аналогия:** Чтобы получить скидку в магазине, вы должны быть `(являетесь студентом) or (у вас есть карта лояльности)`. Достаточно выполнения одного из условий.

````python
is_student = False
has_loyalty_card = True
can_get_discount = is_student or has_loyalty_card
print(f"Студент? {is_student}. Карта есть? {has_loyalty_card}. Будет скидка? -> {can_get_discount}")

is_student = False
has_loyalty_card = False
can_get_discount = is_student or has_loyalty_card
print(f"Студент? {is_student}. Карта есть? {has_loyalty_card}. Будет скидка? -> {can_get_discount}")
````

### Оператор `not` (логическое "НЕ")

Это самый простой оператор. Он просто инвертирует (переворачивает) булево значение. `not True` становится `False`, а `not False` становится `True`.

| A | not A |
| :---: | :---: |
| `True` | `False` |
| `False`| `True` |

**Аналогия:** `is_weekend = True` (сегодня выходной). `need_to_work = not is_weekend` (нужно работать = НЕ выходной).

````python
is_raining = True
print(f"Идет дождь: {is_raining}")

take_umbrella = is_raining
print(f"Взять зонт: {take_umbrella}")

sky_is_clear = not is_raining
print(f"Небо чистое: {sky_is_clear}")
````

## Продвинутые трюки и "подводные камни"

#### Приоритет операторов
Как и в математике (`2 + 2 * 2`), у логических операторов есть порядок выполнения:
1.  Сначала вычисляется `not`.
2.  Затем `and`.
3.  В последнюю очередь `or`.

Если сомневаетесь — используйте скобки `()` для явного указания порядка, как в математике. Это делает код гораздо читабельнее.

#### "Ленивые" вычисления (Short-circuit evaluation)
Это очень важная особенность Python.
- В выражении `A and B`, если `A` оказалось `False`, Python **даже не будет проверять `B`**. Зачем? Результат все равно будет `False`.
- В выражении `A or B`, если `A` оказалось `True`, Python **даже не будет проверять `B`**. Зачем? Результат все равно будет `True`.

Это не только экономит ресурсы, но и позволяет писать безопасный код.

````python
# Пример приоритета
# Как будет вычислено `True or False and False`?
# Сначала `False and False` -> `False`
# Затем `True or False` -> `True`
print(f"True or False and False = {True or False and False}")

# Используем скобки для изменения порядка
# Сначала `True or False` -> `True`
# Затем `True and False` -> `False`
print(f"(True or False) and False = {(True or False) and False}")

print("-" * 20)

# Пример "ленивых" вычислений
# Этот код никогда не вызовет ошибку деления на ноль, потому что до второй части дело не дойдет!
divisor = 0
# `divisor != 0` -> False. Python останавливается и не вычисляет `10 / divisor > 1`.
result = (divisor != 0) and (10 / divisor > 1)
print(f"Безопасная проверка: {result}")
````

## 🔥 Священный Грааль Python: Блоки и Отступы

Прежде чем мы напишем первую конструкцию, нужно усвоить самое главное правило синтаксиса Python, которое отличает его от большинства других языков.

В C++, Java, JavaScript и многих других языках блоки кода, относящиеся к условию, выделяются фигурными скобками `{}`.
```javascript
// Пример на JavaScript
if (age >= 18) {
  // Этот код внутри скобок выполнится, если условие истинно
  console.log("Доступ разрешен");
}

В Python фигурных скобок НЕТ. Вместо них используется ОТСТУП (indentation).

❗️**ПРАВИЛО**: Код, который должен выполниться внутри условия, сдвигается вправо на 4 пробела. Весь код с одинаковым отступом считается одним блоком. 

Это не рекомендация для красоты. Это жесткое синтаксическое правило. Если вы ошибетесь с отступом, программа выдаст ошибку IndentationError или будет работать не так, как вы ожидаете. Ваш редактор кода (VS Code, PyCharm) обычно ставит 4 пробела автоматически, когда вы нажимаете Tab.

## Простейшая конструкция: `if`

Используется, когда нам нужно выполнить какое-то действие **только если** условие истинно. Если оно ложно — программа просто проигнорирует этот блок и пойдет дальше.

## Цепочка проверок: `if-elif-else`



Что если у нас больше двух вариантов? Например, система оценок: "отлично", "хорошо", "удовлетворительно", "неудовлетворительно". Для этого используется `elif` (сокращение от `else if`).

Программа проверяет условия по цепочке, сверху вниз:
1.  Проверяет `if`. Если истинно — выполняет его блок и **выходит из всей конструкции**.
2.  Если `if` ложно, проверяет первый `elif`. Если он истинен — выполняет его блок и **выходит**.
3.  ...и так далее по всем `elif`.
4.  Если ни одно из условий `if` и `elif` не сработало, выполняется блок `else` (он необязателен).

**❗️Важно:** В такой цепочке всегда выполняется **не более одного** блока кода.

````python
temperature = int(input("Введите температуру на улице (в градусах Цельсия): "))

if temperature > 25:
    print("Очень жарко! Наденьте шорты и кэпку.")
elif temperature > 15: # Эта проверка сработает, только если temperature <= 25
    print("Тепло. Джинсы и легкая куртка подойдут.")
elif temperature > 0:  # Эта проверка сработает, только если temperature <= 15
    print("Прохладно. Надевайте теплое пальто и шапку.")
elif temperature > -25: 
    print("Морозно! Нужен пуховик, шарф и перчатки.")
elif temperature > -75: 
    print("Холодно! Шуба и валенки")
else: # Этот блок сработает, если temperature <= -75
    print("Экстремальный холод! Водка замерзла. Грустно. Лучше остаться дома.")
````

Рассмотрим конкретный пример. Допустим, в зависимости от возраста пользователя, ему рекомендуется определенный видеоконтент. При этом выделяют группы от 3 до 6 лет, от 6 до 12, от 12 до 16, 16+. Итого 4 диапазона. Как бы мы стали реализовывать задачу, имея в наборе инструментов только конструкцию if-else?

Самый простой ответ – последовательно проверять вхождение введенного числа-возраста в определенный диапазон с помощью следующих друг за другом условных операторов:


````python
old = int(input('Ваш возраст: '))

print('Рекомендовано:', end=' ')

if 3 <= old < 6:
    print('"Заводной апельсин"')
if 6 <= old < 12:
    print('"Зеленый слоник"')
if 12 <= old < 16:
    print('"Фиксики"')
if 16 <= old:
    print('"Гравити фолз"')
````

Предложенный код прекрасно работает, но есть одно существенное "но". Он не эффективен, так как каждый if в нем – это отдельно взятый оператор, никак не связанный с другими if. Процессор тратит время и "нервы" на обработку каждого из них, даже если в этом уже нет необходимости. Например, введено число 10. В первом if логическое выражение возвращает ложь, и поток выполнения переходит ко второму if. Логическое выражение в его заголовке возвращает истину, и его тело выполняется. Всё, на этом программа должна была остановиться.

Решить проблему избыточности проверок можно, вкладывая условные операторы друг в друга:


````python
old = int(input('Ваш возраст: '))
m=1
print('Рекомендовано:', end=' ')
 
if 3 <= old < 6:
    print('"Петя и Волк"')
    m=m+1
else:
    if 6 <= old < 12:
        print('"Иваново детство"')
        m=m+2
    else:
        if 12 <= old < 16:
            m=m+3
            print('"Восхитительный мистер Фокс"')
            print(m)
        else:
            if 16 <= old:
                print('"Когда плачут Чайки"')
                
print(m)
````

Рассмотрим поток выполнения этого варианта кода. Сначала проверяется условие в первом if (он же самый внешний). Если здесь было получено True, то тело этого if выполняется, а в ветку else мы даже не заходим, так как она срабатывает только тогда, когда в условии if возникает ложь.

Если внешний if вернул False, поток выполнения программы заходит в соответствующий ему внешний else. В его теле находится другой if со своим else. Если введенное число попадает в диапазон от 6 до 12, то выполнится тело вложенного if, после чего программа завершается. Если же число не попадает в диапазон от 6 до 12, то произойдет переход к ветке else. В ее теле находится свой условный оператор, имеющий уже третий уровень вложенности.

Таким образом до последней проверки (16 <= old) интерпретатор доходит только тогда, когда все предыдущие возвращают False. Если же по ходу выполнения программы возникает True, то все последующие проверки опускаются, что экономит ресурсы процессора. Кроме того, такая логика выполнения программы более правильная.

Теперь зададимся следующим вопросом. Можно ли как-то оптимизировать код множественного ветвления и не строить лестницу из вложенных друг в друга условных операторов? Во многих языках программирования, где отступы используются только для удобства чтения программистом, но не имеют никакого синтаксического значения, часто используется подобный стиль:

````python
if логическое_выражение {
    … ;
}
else if логическое_выражение {
    … ;
}
else if логическое_выражение {
    … ;
}
else {
    … ;
}
````

Может показаться, что имеется только один уровень вложенности, и появляется новое расширение для if, выглядящее как else if. Но это только кажется. На самом деле if, стоящее сразу после else, является вложенным в это else. Выше приведенная схема – то же самое, что


````python
if логическое_выражение {
    … ;
}
else
    if логическое_выражение {
        … ;
    }
    else
        if логическое_выражение {
            … ;
        }
        else {
            … ;
        }
````

Именно так ее "понимает" интерпретатор или компилятор. Однако считается, что человеку проще воспринимать первый вариант.

В Питоне такое поднятие вложенного if к внешнему else невозможно, потому что здесь отступы и переходы на новую строку имеют синтаксическое значение. Поэтому в язык Python встроена возможность настоящего множественного ветвления на одном уровне вложенности, которое реализуется с помощью веток elif.

Слово "elif" образовано от двух первых букв слова "else", к которым присоединено слово "if". Это можно перевести как "иначе если".

В отличие от else, в заголовке elif обязательно должно быть логическое выражение также, как в заголовке if. Перепишем нашу программу, используя конструкцию множественного ветвления:

````python
old = int(input('Ваш возраст: '))
 
print('Рекомендовано:', end=' ')
 
if 3 <= old < 6:
    print('"Прощай унылый учитель"')
elif 6 <= old < 12:
    print('"Помпоко: Война тануки в период Хэйсэй"')
elif 12 <= old < 16:
    print('"Ковбой Бибоп"')
elif 16 <= old:
    print('"Триган"')
````

Обратите внимание, в конце, после всех elif, может использоваться одна ветка else для обработки случаев, не попавших в условия ветки if и всех elif. Блок-схему полной конструкции if-elif-…-elif-else можно изобразить так:


Как только тело if или какого-нибудь elif выполняется, программа сразу же возвращается в основную ветку (нижний ярко-голубой прямоугольник), а все нижеследующие elif, а также else пропускаются.


### Ловушка `if-elif`: Порядок имеет значение!

Если поменять условия местами, логика может сломаться. Python выполняет первое же истинное условие и выходит.

````python
# Неправильный порядок проверки оценок
grade = 95

if grade > 50: # 95 > 50 -> True. Это условие сработает первым!
    print("Удовлетворительно.")
elif grade > 70:
    print("Хорошо.")
elif grade > 90:
    print("Отлично!")

# Программа выведет "Удовлетворительно.", хотя оценка 95.
# Правильно всегда идти от более строгих проверок к более общим.
````

## "Синтаксический сахар": Тернарный оператор

# Что такое строка? Последовательность символов

Мы уже знаем, что строка (`str`) — это тип данных для хранения текста. Но есть более точное и мощное определение: **строка — это упорядоченная последовательность символов**.

"Упорядоченная" — это ключевое слово. Оно означает, что каждый символ в строке имеет свой уникальный номер (индекс), и их порядок фиксирован. Это открывает нам огромные возможности для работы с текстом.

**Аналогия:** Представьте, что строка — это улица, а каждый символ — это дом на этой улице. У каждого дома есть свой номер, и мы можем обратиться к любому дому, зная его номер.

## Способы создания строк

Python очень гибок в том, как вы можете определять строки.

- **Одинарные кавычки (`'`):** Самый распространенный способ.

##  Индексы и срезы: "Хирургическая" работа с текстом

Так как строка — это последовательность, мы можем получать доступ к ее частям.

### Индексация: Доступ к одному символу

- Индексация — это получение одного символа по его номеру.
- **❗️ВАЖНО:** Нумерация в Python (как и в большинстве языков программирования) начинается с **НУЛЯ**! Первый элемент имеет индекс `0`, второй — `1`, и так далее.
- **Отрицательные индексы:** Python позволяет вести отсчет с конца. `-1` — это последний элемент, `-2` — предпоследний, и т.д.

````python
#            P  y  t  h  o  n
# Индексы:   0  1  2  3  4  5
# Отрицат.: -6 -5 -4 -3 -2 -1
````

````python
word = "Python"

# Получаем первый символ (индекс 0)
first_char = word[0]
print(f"Первый символ слова '{word}': {first_char}")

# Получаем третий символ (индекс 2)
third_char = word[2]
print(f"Третий символ слова '{word}': {third_char}")

# Получаем последний символ (индекс -1)
last_char = word[-1]
print(f"Последний символ слова '{word}': {last_char}")

# Попытка обратиться к несуществующему индексу вызовет ошибку IndexError
# error_char = word[10]
````

### Срезы (Slices): Получение целого куска строки

Срез позволяет "вырезать" из строки подстроку, указав начальный и конечный индексы.

**Синтаксис:** `строка[start:end:step]`

- `start`: Индекс, с которого начинаем срез (включается в результат). Если опущен, срез начинается с самого начала строки.
- `end`: Индекс, на котором заканчиваем срез (**не включается** в результат). Если опущен, срез идет до самого конца строки.
- `step`: Шаг, с которым идем по строке. Если опущен, по умолчанию равен `1`.

````python
phrase = "Анализ данных"
# Индексы: 0123456789...

# Срез с 0 по 5 символ (6-й не включается)
slice_1 = phrase[0:6]
print(f"Срез [0:6]: '{slice_1}'")

# Если начинаем с 0, `start` можно опустить
slice_2 = phrase[:6]
print(f"Срез [:6]: '{slice_2}'")

# Срез с 7 символа и до конца
slice_3 = phrase[7:]
print(f"Срез [7:]: '{slice_3}'")

# Срез с шагом 2 (берем каждый второй символ)
every_second_char = phrase[::2]
print(f"Каждый второй символ: '{every_second_char}'")

# ❗️Самый известный трюк со срезами: переворот строки
reversed_phrase = phrase[::-1]
print(f"Перевернутая строка: '{reversed_phrase}'")
````

Срез (slice) — извлечение из данной строки одного символа или некоторого фрагмента (подстроки). Есть три формы срезов. Самая простая форма среза: взятие одного символа строки, а именно, S[i] — это срез, состоящий из одного символа, который имеет номер i, при этом считая, что нумерация начинается с числа 0. То есть если S = ‘Hello’, то S[0]=='H', S[1]=='e', S[2]=='l', S[3]=='l', S[4]=='o'.

Номера символов в строке (а также в других структурах данных: списках, кортежах) называются индексом.

##  Методы строк: Встроенные возможности

Как мы уже знаем, строки — это объекты со встроенными "умениями" — методами. Давайте рассмотрим самые полезные из них для аналитика данных.

### Методы для очистки и форматирования:
- `.strip()`: Убирает пробельные символы (пробелы, табы, переносы строк) **в начале и в конце** строки. Это **must-have** для очистки данных, полученных из файлов или от пользователей.
- `.lower()` / `.upper()`: Приводят всю строку к нижнему или верхнему регистру. Необходимо, чтобы унифицировать данные. Например, чтобы "Россия", "россия" и "РОССИЯ" считались одним и тем же.
- `.replace(old, new)`: Заменяет все вхождения подстроки `old` на `new`.
- `.capitalize()` / `.title()`: Делает заглавной первую букву всей строки / каждого слова.

### Методы-трансформеры (супер-важные!):
- `.split(separator)`: **Разделяет** одну строку на **список** из нескольких строк по указанному разделителю. Если разделитель не указан, разделяет по пробелам.
- `.join(list_of_strings)`: **Собирает** из списка строк одну-единственную строку, вставляя между элементами строку-разделитель, у которой был вызван метод.

````python
# --- Очистка и форматирование ---
messy_data = "   Москва, Россия   \n"
print(f"Исходные данные: '{messy_data}'")

# 1. Убираем лишние пробелы и переносы строк
clean_data = messy_data.strip()
print(f"После .strip(): '{clean_data}'")

# 2. Приводим к единому регистру
lower_case_data = clean_data.lower()
print(f"После .lower(): '{lower_case_data}'")

# 3. Заменяем запятую на дефис
replaced_data = lower_case_data.replace(",", " -")
print(f"После .replace(): '{replaced_data}'")
print("-" * 30)

# --- Трансформация: split и join ---

# Задача: из строки "яблоко,банан,вишня" получить список ['яблоко', 'банан', 'вишня']
csv_string = "яблоко,банан,вишня"
fruits_list = csv_string.split(',') # Разделяем по запятой
print(f"Результат .split(','): {fruits_list}")
print(f"Первый элемент списка: {fruits_list[0]}")

# Обратная задача: из списка `fruits_list` собрать строку, разделенную пробелами
fruits_to_join = ['яблоко', 'банан', 'вишня']
# ' ' - это строка-разделитель. Мы вызываем у нее метод .join()
joined_string = " ".join(fruits_to_join)
print(f"Результат ' '.join(...): '{joined_string}'")
````

# Концепция циклов `while`: День сурка в программировании

Представьте фильм "День сурка", где герой проживал один и тот же день снова и снова. Циклы в программировании — это наш способ создать такой "день сурка" для компьютера.

**Цикл** — это управляющая конструкция, которая позволяет выполнять один и тот же блок кода многократно, пока выполняется некоторое условие.

До сих пор наши программы были строго линейными: выполнили команду A, затем B, затем C и завершились. Циклы позволяют нам сказать: "Выполняй блок команд X, пока не наступит событие Y". Это основа любой автоматизации. Вместо того чтобы вручную обрабатывать 1000 файлов, мы пишем цикл, который сделает это за нас.

##  Анатомия цикла `while`

Цикл `while` (в переводе "пока") — самый базовый и универсальный тип цикла. Его логика проста: "ПОКА условие истинно (`True`), ДЕЛАЙ...".

**Классическая структура цикла `while` состоит из четырех обязательных частей:**

1.  **Инициализация:** Мы создаем переменную-"счетчик" или "флаг" **до** начала цикла. Она будет управлять его работой.
2.  **Условие:** После ключевого слова `while` мы пишем условие, которое проверяется **перед каждой** итерацией (повторением) цикла.
3.  **Тело цикла:** Блок кода с отступом, который будет выполняться, если условие истинно.
4.  **Изменение:** Внутри тела цикла мы **обязательно** должны сделать что-то, что в конечном итоге повлияет на условие и сделает его ложным. Чаще всего это изменение переменной-"счетчика".

Давайте посмотрим на простейший пример: печать чисел от 1 до 5.

````python
# Пример: Считаем от 1 до 5

# 1. Инициализация
counter = 1

# 2. Условие
while counter <= 5:
    # 3. Тело цикла
    print('Цикл выполнился', counter, 'раз(а)')
    
    # 4. Изменение
    counter += 1 # Эквивалентно counter = counter + 1

# Этот код выполнится только после того, как условие цикла станет False
print("Цикл завершил свою работу.")
````

##  Главная опасность: Бесконечный цикл

Что произойдет, если мы забудем шаг №4 (Изменение)? Условие `counter <= 5` будет истинно **всегда** (`1 <= 5` -> `True`), и цикл никогда не остановится. Это называется **бесконечный цикл**.

> **❗️ВАЖНО:** Бесконечный цикл — это не ошибка, которую отловит Python. Программа просто "зависнет", выполняя одно и то же действие вечно (или пока вы ее принудительно не остановите).

##  Практические примеры использования `while`

Циклы `while` идеальны для ситуаций, когда мы не знаем заранее, сколько раз нужно будет повторить действие.

### Пример 1: Валидация пользовательского ввода

Мы хотим, чтобы пользователь ввел свой возраст, и нам нужно убедиться, что это положительное число. Мы будем спрашивать его снова и снова, **пока** он не введет корректное значение.

````python
age = int(input("Введите ваш возраст: "))

while age <= 0:
    print("Ошибка! Возраст должен быть положительным числом.")
    age = int(input("Пожалуйста, попробуйте снова. Введите ваш возраст: "))

print(f"Спасибо! Ваш возраст ({age}) успешно записан.")
````

### Пример 2: Простой интерактивный "диалог"

Мы можем создать программу, которая будет "общаться" с пользователем, пока он сам не решит закончить диалог, введя стоп-слово.

````python
user_input = "" # Инициализируем пустой строкой

while user_input.lower() != "выход":
    user_input = input("Скажите что-нибудь (или напишите 'выход', чтобы закончить): ")
    
    # Проверим, чтобы не отвечать на само слово "выход"
    if user_input.lower() != "выход":
        print(f"Вы сказали: '{user_input}'")

print("Программа завершена. До свидания!")
````

##  Управляем циклом: `break` и `continue`

Иногда нам нужно более тонко управлять поведением цикла: выйти из него досрочно или пропустить текущую итерацию.

### `break`: Аварийный выход

Оператор `break` **немедленно и полностью** прекращает выполнение цикла, даже если его условие все еще истинно. Программа переходит к первой строке кода **после** цикла.

**Аналогия:** Это как "стоп-кран" в поезде.

````python
# Перепишем наш "диалог" с использованием `while True` и `break`
# `while True` - это самый простой способ создать бесконечный цикл намеренно.
# Выход из такого цикла возможен ТОЛЬКО через `break`. Это очень распространенная идиома.

print("--- Версия диалога с break ---")
while True:
    user_input = input("Скажите что-нибудь (или напишите 'выход', чтобы закончить): ")
    
    if user_input.lower() == "выход":
        break # Немедленно выходим из цикла
    
    print(f"Вы сказали: '{user_input}'")

print("Программа завершена. До свидания!")
````

### `continue`: Пропустить итерацию

Оператор `continue` **прекращает текущую итерацию** и немедленно переходит к **проверке условия** для следующей итерации. Код, который идет после `continue` в теле цикла, не выполняется.

**Аналогия:** Это как кнопка "пропустить ход" в настольной игре.

````python
# Задача: напечатать все нечетные числа от 1 до 10.
number = 0
while number < 10:
    number += 1
    
    # Если число четное (остаток от деления на 2 равен 0)...
    if number % 2 == 0:
        continue # ...пропускаем оставшуюся часть цикла (print) и идем на следующую итерацию
    
    print(f"Нечетное число: {number}")
````

## Необязательный блок: `else` в цикле `while`

У цикла `while` в Python есть одна необычная особенность — опциональный блок `else`.

Блок `else` выполняется только в том случае, если цикл завершился **естественным образом** (то есть его условие стало `False`). Если же цикл был прерван оператором `break`, блок `else` **не выполняется**.

Это бывает полезно в задачах поиска: если мы нашли элемент и вышли через `break`, `else` не сработает. Если же мы прошли весь цикл и ничего не нашли, `else` сработает, и мы сможем сообщить "Элемент не найден".

````python
# Пример: Поиск "плохого" числа в последовательности
counter = 1
while counter <= 5:
    print(f"Проверяем число {counter}...")
    if counter == 6: # Этого числа нет в нашей последовательности
        print("Найдено 'плохое' число! Прерываем.")
        break
    counter += 1
else:
    # Этот блок сработает, т.к. break не был вызван
    print("Цикл завершился нормально, 'плохих' чисел не найдено.")
print("Циклы закончены")
print("-" * 30)

# А теперь найдем "плохое" число
counter = 1
while counter <= 5:
    print(f"Проверяем число {counter}...")
    if counter == 3: # А это число есть
        print("Найдено 'плохое' число! Прерываем.")
        break
    counter += 1
else:
    # Этот блок НЕ сработает, т.к. цикл был прерван
    print("Цикл завершился нормально, 'плохих' чисел не найдено.")
print("Циклы закончены")
````

##  Концепция списка: От одной коробки к стеллажу

До сих пор мы работали с переменными, которые были как одиночные "коробки" для хранения одного значения (`age = 25`, `name = "Анна"`). Этого катастрофически мало для реальных задач.

**Список (`list`)** — это структура данных, которую можно представить как **стеллаж с пронумерованными ячейками**. В каждую ячейку мы можем положить что угодно, и у каждой ячейки есть свой уникальный номер (индекс).

**Ключевые свойства списков:**
- **Упорядоченность:** Элементы хранятся в том порядке, в котором вы их добавили.
- **Изменяемость (Mutability):** Вы можете изменять, добавлять и удалять элементы из списка **после** его создания. Это их главное отличие от строк!
- **Гибкость:** Список может хранить элементы абсолютно разных типов данных одновременно (хотя чаще всего хранят однотипные).

Список - это набор элементов в определенном порядке. Вы можете составить список, включающий буквы алфавита, цифры от 0 до 9 или имена всех членов вашей семьи. Вы можете поместить в список все, что хотите, и элементы в вашем списке не обязательно должны быть связаны каким-либо определенным образом. Поскольку список обычно содержит более одного элемента, рекомендуется сделать имя вашего списка множественным числом, например, буквами, цифрами или именами.

##  Доступ к элементам: Индексы и срезы

Работа с индексами и срезами в списках **абсолютно идентична** тому, как мы это делали со строками. Это одна из прекрасных особенностей Python — единый подход к последовательностям.

````python
#            0         1        2       3        4
planets = ["Меркурий", "Венера", "Земля", "Марс", "Юпитер"]
# Отрицат.:   -5        -4       -3      -2       -1

# Доступ по индексу
first_planet = planets[0]
print(f"Первая планета от Солнца: {first_planet}")

third_planet = planets[2]
print(f"Мы живем на планете: {third_planet}")

last_planet = planets[-1]
print(f"Самая большая планета: {last_planet}")

# Срезы
inner_planets = planets[0:4] # 4-й элемент (Марс) включен, 5-й (Юпитер) - нет
print(f"Планеты земной группы: {inner_planets}")

gas_giants = planets[4:] # С 4-го индекса и до конца
print(f"Газовые гиганты (в нашем списке): {gas_giants}")
````

##  Изменяемость (Mutability)

Вспомните, строки были **иммутабельными** (неизменяемыми). Мы не могли поменять один символ в строке, только создать новую. Списки — **мутабельны** (изменяемы). Мы можем "залезть" в любую ячейку по ее индексу и заменить ее содержимое.

````python
my_pets = ["кот", "собака", "хомяк"]
print(f"Мои питомцы (изначально): {my_pets}")

# Давайте заменим попугая на хомяка
my_pets[2] = "игуана" # Обращаемся по индексу и присваиваем новое значение
print(f"Мои питомцы (после смерти хомяка): {my_pets}") # Исходный список изменился!
````

#  Концепция: `for` против `while`

Мы уже знаем цикл `while`, который работает "пока условие истинно". Он универсален, но часто бывает избыточен.

Цикл `for` создан специально для одной задачи: **пройтись по каждому элементу в последовательности** (например, в списке или строке) и что-то с ним сделать.

| `while` | `for` |
| :--- | :--- |
| Работает по **условию** | Работает по **последовательности** |
| "Делай, ПОКА голоден" | "Съешь **КАЖДЫЙ** пельмень в тарелке" |
| Нужно вручную управлять счетчиком | Сам следит за перебором элементов |

В 90% случаев, когда вам нужно перебрать элементы списка, **цикл `for` — это более простой, безопасный и "питонический" (Pythonic) способ**.

## Базовый синтаксис

**Синтаксис:**
```python
for временная_переменная in последовательность:

## Добавление `if` в цикл

Мы можем добавить if в начало цикла, чтобы проверить условия перед продолжением. Вы можете поместить любой условный тест в первую строку и практически любое действие в блоке с отступом после теста. Если условный тест оценивается как True, Python выполняет код, следующий за оператором if. Если результат теста равен False, Python игнорирует код, следующий за оператором if.

## Функция `range()`: Цикл на заданное число повторений

Что если мы хотим повторить действие N раз, но у нас нет готового списка? Для этого существует функция `range()`. Она генерирует последовательность чисел "на лету".

- `range(stop)`: Генерирует числа от `0` до `stop-1`.
- `range(start, stop)`: Генерирует числа от `start` до `stop-1`.
- `range(start, stop, step)`: Генерирует числа от `start` до `stop-1` с шагом `step`.

## `enumerate()`: Получаем и индекс, и значение

Часто при переборе списка нам нужен не только сам элемент, но и его порядковый номер (индекс).

**Плохой способ (в стиле языка C):**
```python
for i in range(len(my_list)):
    print(f"Индекс: {i}, Значение: {my_list[i]}")

Хороший, **"питонячий"** способ: использовать функцию `enumerate()`. Она превращает список ['a', 'b'] в последовательность пар (0, 'a'), (1, 'b').



````python
tasks = ["Сходить на работу/отправить резюме", "Сходить на пару к Воробьеву Д.А.", "Лечь спать пораньше"]

# `enumerate` сразу дает нам и индекс, и значение
for index, task in enumerate(tasks):
    # По умолчанию нумерация с 0, добавим 1 для красоты
    print(f"Задача №{index + 1}: {task}")
````

## `break` и `continue` в цикле `for`

Операторы `break` и `continue` работают в цикле `for` **абсолютно так же**, как и в `while`.
- `break`: немедленно прерывает цикл.
- `continue`: пропускает текущую итерацию и переходит к следующему элементу.

````python
# Ищем первое число больше 5 в списке
numbers = [1, 3, 4, 8, 2, 9, 7]
for num in numbers:
    if num > 5:
        print(f"Найдено первое число больше 5: {num}")
        break # Цикл прекращает работу
````

````python
# Печатаем все имена, кроме "Анна"
names = ["Иван", "Мария", "Анна", "Петр"]
for name in names:
    if name == "Анна":
        continue # Пропускаем Анну и идем дальше
    print(f"Привет, {name}!")
````

## Еще немного питонячества


Решим простую задачку со списками, заодно немного посмотрев на то, как работают всякие циклы и ветвления в этом нашем питоне.

Допустим, у нас есть список чисел и мы хотим его "пофильтровать", оставив только чётные элементы этого списка, сохранив его в новом списке. Есть несколько способов это сделать.

````python
initial = [2, 4, 7335, 753, 2, -17, -22]
````

###### Способ 1
Это как бы мы сделали в каком-нибудь простом языке типа C с минимумом примочек.


````python
odds1 = []  # пустой список
i = 0
while i < len(initial):
    current = initial[i]
    current_is_odd = ((current % 2) == 1)
    if current_is_odd:
        odds1.append(current)
        
    i = i + 1
````

###### Способ 2


Немножко более питонячий, мы, избавляемя от явного i = 0 и i = i + 1, за счёт этого чуть меньше вероятность ошибиться

````python
odds2 = []

for current in initial:
    current_is_odd = ((current % 2) == 1)
    if current_is_odd:
        odds2.append(current)
````

###### Способ 3
Наиболее питонячий, прикольный, однострочный и молодёжный

````python
odds3 = [current for current in initial if (current % 2 == 1)]
````

````python
print("Способ 1:", odds1)
print("Способ 2:", odds2)
print("Способ 3:", odds3)
````
