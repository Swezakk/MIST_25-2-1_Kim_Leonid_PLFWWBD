````python
def mega_function(*args, **kwargs):
    print("---   (*args) ---")
    if args:
        for arg in args:
            print(arg)
    else:
        print("  .")
        
    print("\n---   (**kwargs) ---")
    if kwargs:
        for key, value in kwargs.items():
            print(f"{key} -> {value}")
    else:
        print("  .")
````

````python
class Student:
    """,  ."""
    
    def __init__(self, name, course):
        self.name = name
        self.course = course
        self.grades = []

    #    
    def add_grade(self, grade):
        """     ."""
        self.grades.append(grade)
        print(f" {self.name}   {grade}")

    #     
    def get_average_grade(self):
        """   ."""
        if not self.grades:
            return 0.0
        return sum(self.grades) / len(self.grades)

    #    
    def display_info(self):
        """    ."""
        print(f": {self.name}, : {self.course}")
        print(f": {self.grades}")
        print(f" : {self.get_average_grade():.2f}")


# ---    ---
maria = Student("", 1)

#   
maria.display_info()
print("-" * 20)
maria.add_grade(5)
maria.add_grade(4)
maria.add_grade(5)
print("-" * 20)
maria.display_info()
````

````python
#  () 
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"  {self.name},  {self.age} .")

# -.    .
class Student(Person):
    def __init__(self, name, age, course):
        #    ,      (name, age)
        super().__init__(name, age) 
        #   ,  
        self.course = course

    #   ,    
    def introduce(self):
        print(f"  {self.name},  {self.age} ,   {self.course} .")

#   -
class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    #   
    def introduce(self):
        print(f"  {self.name},  {self.age} .   {self.subject}.")

p = Person("", 30)
s = Student("", 20, 2)
t = Teacher(" ", 45, "Python")

p.introduce()
s.introduce() #     Student
t.introduce() #     Teacher
````

````python
class NutritionInfo:
    def __init__(self, proteins, carbs, fats):
        self.proteins = proteins
        self.carbs = carbs
        self.fats = fats
    
    def __add__(self, other):
        return NutritionInfo(self.proteins + other.proteins, self.carbs + other.carbs, self.fats + other.fats)
    
    def __str__(self):
        return f"Nutrition p {self.proteins}, c {self.carbs}, f {self.fats}"
        
    def energy(self):
        return int(self.fats * 9 + (self.carbs + self.proteins) * 4.2)
````

````python
import time

def timer_decorator(func_to_decorate):
    #  -""     
    # *args  **kwargs ,      
    def wrapper(*args, **kwargs):
        print(f"---   '{func_to_decorate.__name__}'... ---")
        start_time = time.time()
        
        #       
        result = func_to_decorate(*args, **kwargs)
        
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"---  '{func_to_decorate.__name__}'   {execution_time:.4f} . ---")
        
        #    ,    
        return result
    
    #    -
    return wrapper

@timer_decorator
def long_running_function(n):
    """   - ."""
    total = 0
    for i in range(n):
        total += i
    print(" .")
    return total

result = long_running_function(10000000)
print(f" : {result}")
````

````python
def benchmark(func):
    import time
    
    def wrapper(*args, **kwargs):
        start = time.time()
        return_value = func(*args, **kwargs)
        end = time.time()
        print('[*]  : {} .'.format(end-start))
        return return_value
    return wrapper

@benchmark
def fetch_webpage(url):
    import requests
    webpage = requests.get(url)
    return webpage.text

webpage = fetch_webpage('https://ya.ru')
print(webpage)
````

````python
def benchmark(iters):
    def actual_decorator(func):
        import time
        
        def wrapper(*args, **kwargs):
            total = 0
            for i in range(iters):
                start = time.time()
                return_value = func(*args, **kwargs)
                end = time.time()
                total = total + (end-start)
            print('[*]   : {} .'.format(total/iters))
            return return_value

        return wrapper
    return actual_decorator


@benchmark(iters=10)
def fetch_webpage(url):
    import requests
    webpage = requests.get(url)
    return webpage.text

webpage = fetch_webpage('http://info.cern.ch')
print(webpage)
````

````python
def memoize_func(f):
    memo = dict()
    def func(*args):
        print(f'Run with args={args}, memo={memo}')
        if args not in memo:
            memo[args] = f(*args)
        return memo[args]

    return func
````

````python
def benchmark(func):
    import time
    
    def wrapper(*args, **kwargs):
        start = time.time()
        return_value = func(*args, **kwargs)
        end = time.time()
        print('[*]  : {} .'.format(end-start))
        return return_value
    return wrapper
@memoize_func

def func(a, b):
    print(f'    Run func({a}, {b})')
    return a ** b
````
